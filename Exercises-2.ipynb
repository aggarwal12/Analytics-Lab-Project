```python
%matplotlib inline
import pandas as pd
pd.__version__
```

```python
titles = pd.DataFrame.from_csv('titles.csv', index_col=None)
```

```python
cast = pd.DataFrame.from_csv('cast.csv', index_col=None)
cast.head()
```

### What are the ten most common movie names of all time?

```python
all_movies=titles.sort_values(by='title',ascending=False).title.value_counts()
all_movies.head(n=10)
#or
all_movies[:10]
```

### Which three years of the 1930s saw the most films released?

```python
title_1930=titles.year//10*10==1930
```

```python
titles[(title_1930)&(titles.title.sort_values(ascending=False))].year.value_counts()[:3]
```

### Plot the number of films that have been released each decade over the history of cinema.

```python
titles['decade_start'] = (titles['year'] // 10) * 10 + 1
#titles.groupby(titles.title).decade_start
title_decade=titles.decade_start.value_counts()
title_decade.plot(kind='bar',legend=True)
```

### Plot the number of "Hamlet" films made each decade.

```python
title_hamlet=titles.title.str.lower().str.contains('hamlet')
plot_hamlet=titles[title_hamlet].decade_start.value_counts()
plot_hamlet.plot(kind='bar')
```
