```python
release_dates = pd.DataFrame.from_csv('D:/Analytics Lab/Class Work/Day 7/data/release_dates.csv', index_col=None,
                                      parse_dates=['date'], infer_datetime_format=True)
release_dates.head()
```

### Make a bar plot of the months in which movies with "Christmas" in their title tend to be released in the USA.

```python
release_dates.dtypes
```

```python
months = release_dates.date.dt.month
release_dates = release_dates.assign(months=months)

import calendar
release_dates['months'] = release_dates['months'].apply(lambda x: calendar.month_abbr[x])
release_dates.head()
titles=release_dates[release_dates.title.str.contains('Christmas')&release_dates.country.str.contains('USA')].groupby('months')['title'].count().sort_values(ascending=False)
titles.plot(kind="bar")
#plt.bar(titles.months)
```

### Make a bar plot of the months in which movies whose titles start with "The Hobbit" are released in the USA.

```python
titles=release_dates[release_dates.title.str.contains('The Hobbit')&release_dates.country.str.contains('USA')].groupby('months')['title'].count().sort_values(ascending=False)
titles.plot(kind="bar")
```

### Make a bar plot of the day of the week on which movies with "Romance" in their title tend to be released in the USA.

```python
day = release_dates.date.dt.dayofweek
release_dates = release_dates.assign(day=day)

import calendar
release_dates['day'] = release_dates['day'].apply(lambda x: calendar.day_abbr[x])
release_dates.head()
titles=release_dates[release_dates.title.str.contains('Romance')&release_dates.country.str.contains('USA')].groupby('day')['title'].count().sort_values(ascending=False)
titles.plot(kind="bar")
```

### Make a bar plot of the day of the week on which movies with "Action" in their title tend to be released in the USA.

```python
titles=release_dates[release_dates.title.str.contains('Action')&release_dates.country.str.contains('USA')].groupby('day')['title'].count().sort_values(ascending=False)
titles.plot(kind="bar")
```
