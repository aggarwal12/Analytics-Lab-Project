### Challenge: first combine these sales together into a single dataframe, then compute how much money consumers spent on each book in each currency.

```python
france_data = sales2[21:23]

france_data['Units sold'] = france_data['Units sold'].astype(str).astype(float)
france_data['List price'] = france_data['List price'].astype(str).astype(float)

#france_data['currency'] = 'EUR'
#france_data['amount_spent'] = france_data['Units sold'].astype(str).astype(float) * france_data['List price'].astype(str).astype(float)

france_data
```

```python
sales_new = sales1.rename(index=str, columns={"Book title": "Title","Number sold":"Units sold","Sales price":"List price","Royalty paid":"Royalty"})

print(sales_new)
print(sales_new.dtypes)
```

```python
sales_new['Title'] = sales_new['Title'].map({
    'The Bricklayer’s Bible' :"The Bricklayer's Bible", 
    'Pining For The Fisheries of Yore' : 'Pining for the Fisheries of Yore',
    'The Tower Commission Report': 'The Tower Commission Report',
    'The Duck Goes Here':'The Duck Goes Here',
    'Swimrand' : 'Swimrand'
})
```

```python
france_data = pd.concat([france_data,sales_new]).drop_duplicates(['Title']).reset_index(drop=True)
uk_data = pd.concat([uk_data,sales_new]).drop_duplicates(['Title']).reset_index(drop=True)
us_data = pd.concat([us_data,sales_new]).drop_duplicates(['Title']).reset_index(drop=True)
```

```python
print(france_data)
print(us_data)
print(uk_data)
```

```python
us_data['currency'] = 'USD'
us_data['amount_spent'] = us_data['Units sold'].astype(str).astype(float) * us_data['List price'].astype(str).astype(float)

france_data['currency'] = 'EUR'
france_data['amount_spent'] = france_data['Units sold'].astype(str).astype(float) * france_data['List price'].astype(str).astype(float)

uk_data['currency'] = 'GBP'
uk_data['amount_spent'] = uk_data['Units sold'].astype(str).astype(float) * uk_data['List price'].astype(str).astype(float)

print(france_data)
print(us_data)
print(uk_data)
```

```python
final_dataframe = pd.concat([us_data, uk_data, france_data]).reset_index(drop=True)

final_dataframe[['Title', 'currency', 'amount_spent']].sort_values(by='Title').reset_index().drop('index', axis=1)
```
